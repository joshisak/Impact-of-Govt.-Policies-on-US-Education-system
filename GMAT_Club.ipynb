```python
import pandas as pd
import numpy as np
#import gensim
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import *
#import numpy as np
#np.random.seed(2018)
import nltk
#from nltk import ngrams
from nltk.corpus import stopwords
import re
import string
from gensim import corpora
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import words
#import collections
```

```python
def clean(doc):
    stop = set(stopwords.words('english'))
    exclude = set(string.punctuation) 
    lemma = WordNetLemmatizer()
    text = words.words()
    wordlist = [w for w in text if 2 < len(w) < 11]
    
    num_free = re.sub('[^a-z ]',' ',str(doc.lower()))
    len_word = " ".join(word for word in num_free.split() if 2 < len(word) < 11)
    stop_free = " ".join([i for i in len_word.split() if i not in stop])
    normalized = " ".join(lemma.lemmatize(word) for word in stop_free.split())
    output = " ".join([i for i in normalized.split() if i in wordlist])

    return output
```

```python
cleaned_text = []
i = 0
for row in df['Text']:
    print(i)
    i = i + 1
    cleaned_row = clean(row)
    cleaned_text.append(cleaned_row)
clean_text = pd.DataFrame(cleaned_text,columns=["Clean Text"])
#df_clean = df_clean.append(clean_text)
```

```python
df = pd.read_csv("2020 MBA forum data.csv")
```

```python
cleaned_text = list(df['Clean_Text'])
sep = ' '
all_text = sep.join(cleaned_text)
all_text = all_text.split()

from nltk import ngrams, FreqDist
all_counts = FreqDist(ngrams(all_text, 2))

ngram_list = []
freq = []
for k,v in all_counts.items():
    ngram_list.append(k)
    freq.append(v)
ngram_list = pd.DataFrame(ngram_list,columns=["Word_1","Word_2"])
freq = pd.DataFrame(freq,columns=["Frequency"])
Ngram = pd.concat([ngram_list, freq], axis=1, sort=False)
result = Ngram.sort_values(by=['Frequency'], ascending=False)
result
```

```python
result.to_csv("bigrams.csv")
```
